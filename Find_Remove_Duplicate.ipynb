*---------------------Find and Remove Duplicate---------------------------*

```python
import pandas as pd
```

```python
df = pd.read_table('./data2/u.user',sep='|',header=None)
```

```python
df.head()
```

```python
df.rename(
    columns={
        0:'User_Id',1:'Age',2:'Sex',3:'Occupation',4:'Zip_Code'},inplace=True
)
```

```python
df.set_index('User_Id',inplace=True)
```

```python
df.head()
```

```python
df.loc[df.duplicated(), : ]       # Show row which are duplicated
```

```python
df.loc[df.duplicated(keep=False), : ] 
```

```python
df.drop_duplicates(keep='first').shape       # keep='last'
```

```python
# Consider column to be unique
df.duplicated(subset=['Age','Zip_Code']).sum()
```
